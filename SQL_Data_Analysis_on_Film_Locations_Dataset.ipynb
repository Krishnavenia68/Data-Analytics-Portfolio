```python
query = """
SELECT name
FROM sqlite_master
WHERE type ='table';
"""
pd.read_sql_query(query,conn)
```

```python
query = """
SELECT *
FROM film_locations;
"""

pd.read_sql_query(query, conn).head()
```

```python
query = """
SELECT locations,writer
FROM film_locations;
"""
pd.read_sql_query(query,conn).tail()
```

```python
query = """
SELECT "Releaseyear"
FROM film_locations
WHERE"Releaseyear" >2010;
"""
pd.read_sql_query(query,conn)
```

```python
query = """
SELECT Title,"Release year"
FROM film_locations
ORDER BY "Release year"DESc;
"""
pd.read_sql_query(query,conn).head()
```

```python
query ="""
SELECT *
FROM film_locations
LIMIT 10;
"""
pd.read_sql_query(query,conn)
```

```python
query = """
SELECT COUNT(*) AS total_movies
FROM film_locations;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT Director,
COUNT(*) AS total_movies
FROM film_locations
GROUP BY Director
ORDER BY total_movies DESC;
"""

pd.read_sql_query(query, conn).head()
```

```python
query = """
SELECT Director,
COUNT(*) AS total_movies
FROM film_locations
GROUP BY Director
HAVING total_movies > 5
ORDER BY total_movies DESC;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT Title, Locations
FROM film_locations
WHERE Locations LIKE '%Bridge%';
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT *
FROM film_locations
WHERE Director IS NULL;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT AVG("Release Year") AS average_year
FROM film_locations;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT MAX("Release Year") AS latest_movie
FROM film_locations;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT MIN("Release Year") AS oldest_movie
FROM film_locations;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT Title, Director, "Release Year"
FROM film_locations
WHERE "Release Year" > 2000
AND Director IS NOT NULL;
"""

pd.read_sql_query(query, conn)
```
